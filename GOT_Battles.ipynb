```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/battles.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
print('Attacker king counts')
print(df['attacker_king'].value_counts())
print('\nDefender king counts')
print(df['defender_king'].value_counts())
```

### Attacker King vs. Attacker Army


```python
plt.figure(figsize=(10,8))
sns.barplot(x='attacker_king', y='attacker_size', data=df)
```

### Defender King vs. Defender Army

```python
plt.figure(figsize=(10,8))
sns.barplot(x='defender_king', y='defender_size', data=df)
```

```python
plt.figure(figsize=(20,5))
sns.countplot(x='attacker_king', hue='battle_type', data=df)
plt.show()
```
